```python
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
def draw_keypoints(image, outputs, gt, color):
    image = image.detach().cpu()
    outputs = outputs.detach().cpu().numpy()
    gt = gt.detach().cpu().numpy()
    print(outputs.shape,gt.shape)
    
    img = image[0]
    output_keypoint = outputs[0]
    #ground_truth = gt[0]
    img = np.array(img, dtype='float32')
    img = np.transpose(img, (1, 2, 0))

    plt.imshow(img)
    output_keypoint = output_keypoint.reshape(-1, 2)
    for p in range(output_keypoint.shape[0]):
        plt.plot(output_keypoint[p, 0], output_keypoint[p, 1], color)
    plt.axis('off')
    plt.show()
    plt.close()
    return 
```

```python
from dataset import FacialKeypointsDataset
import config
import torch
from torch.utils.data import DataLoader
data=FacialKeypointsDataset(f'{config.ROOT_PATH}/test_frames_keypoints.csv',f'{config.ROOT_PATH}/test')
dataloader=DataLoader(data,batch_size=config.BATCH_SIZE,shuffle=False)
```

```python
from model import FaceKeypointResNet50
model_original = FaceKeypointResNet50(pretrained=False, requires_grad=False).to(config.DEVICE)
checkpoint = torch.load('../outputs/model.pth')
model_original.load_state_dict(checkpoint['model_state_dict'])


model_halfp = FaceKeypointResNet50(pretrained=False, requires_grad=False).to(config.DEVICE)
checkpoint = torch.load('../outputs/model.pth')
model_halfp.load_state_dict(checkpoint['model_state_dict'])
model_halfp=model_halfp.to(torch.float16)


model_prune = FaceKeypointResNet50(pretrained=False, requires_grad=False,pruning_amount = 0.2).to(config.DEVICE)
checkpoint = torch.load('../outputs/model_prune_0.2_new.pth')
model_prune.load_state_dict(checkpoint['model_state_dict'])

model_original.eval()
model_halfp.eval()
model_prune.eval()
```

```python
with torch.no_grad():
    num_batches=int(len(data)/dataloader.batch_size)
    for i,data in tqdm(enumerate(dataloader),total=num_batches):
        image,keypoints=data['image'].to(config.DEVICE),data['keypoints'].to(config.DEVICE)
        keypoints=keypoints.view(keypoints.size(0),-1)
        print('1')
        outputs=model_original(image)
        print('2')
        if i==0:
            draw_keypoints(image,keypoints,keypoints,'b.')
            print('3')
            break
```
